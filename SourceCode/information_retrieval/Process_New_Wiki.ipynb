```python
import os
import json
import pandas as pd
import math

page_identifier_list = []
sentence_number_list = []
text_list = []

def loadfile(folder): 
    a = 0
    list_of_files = os.listdir(folder)
    print(list_of_files)
    
    for file in list_of_files:
        filename = os.path.join(folder, file)
        with open(filename, 'r') as doc:
            a += 1
            print(a)
            lines = doc.readlines()
            for i in range(len(lines)):
#                 print(i)
                words = lines[i].split()
    
                page_identifier_list.append(words[0])
                sentence_number_list.append(words[1])
                text_list.append(lines[i])
#                 corpus_df.loc[i] = [words[0], words[1], " ".join(words[2:])]
#                 corpus_df.loc[i] = [lines[i].split()[0], lines[i].split()[1], " ".join(lines[i].split()[2:])]  
#     return corpus_df

loadfile("new-wiki-pages-text/")
final_data = {
    "page_identifier": page_identifier_list,
    "sentence_number": sentence_number_list,
    "text": text_list
}
corpus_df = pd.DataFrame(final_data)
# corpus_df = loadfile("new-wiki-pages-text")
corpus_df.to_csv('new_wiki.csv',index = False)
```
