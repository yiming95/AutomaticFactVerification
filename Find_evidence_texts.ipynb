```python
import os
import pandas as pd
import json
import linecache
import ast

wiki_file_path = "./new-wiki-pages-text/"
tsv_file_path = "./bert_train_527.csv"
tsv_output_path = "./bert_train_527_result.csv"

df = pd.read_csv(tsv_file_path, sep='\t')

# create index for evidences
evi_index = {}
for file in os.listdir(wiki_file_path):
    print("processing file: " + file)
    file_name = wiki_file_path + file
    with open(file=file_name) as f:
        lines = f.readlines()
        for i in range(len(lines)):
            words = lines[i].split()
            try:
                evi = (words[0], int(words[1]))
                evi_index.update({
                evi: (file, i)
                })
            except ValueError as e:
                pass
```

```python
df = pd.read_csv(tsv_file_path, sep='\t')

evidence_text_list = []
evidence_index_list = list(df['evidence'].values)
for evidence_index in evidence_index_list:
    combination = tuple(ast.literal_eval(evidence_index))
    file_index_tuple = evi_index.get(combination)
    file = wiki_file_path + file_index_tuple[0]
    line_num = file_index_tuple[1]
    evi_text = linecache.getline(file,line_num + 1)
    words = evi_text.split(" ")[2: len(evi_text)]
    words = words[0: -1]
    evi_text = " ".join(words)
    evidence_text_list.append(evi_text)
    
df['evidence_text'] = evidence_text_list

df.to_csv(tsv_output_path, sep='\t', index=False)

df
```
